```julia
h = 0.25
y = ones(5)
for i = 2:5
    y[i] = y[i-1]*(1+h)
end
plot(exp, 0, 1)
plot!(0:h:1, y, c=:red)
```

```julia
function euler(F, x0, y0, xf; n = 10)
    h = (xf - x0)/n
    x = linspace(x0, xf, n+1)
    y = zeros(n+1)
    y[1] = y0
    for i = 1:n
        y[i+1] = y[i] + h * F(x[i], y[i])
    end
    return x, y
end
```

```julia
F(x,y) = y
sol(x) = exp(x)
x, y = euler(F, 0, 1, 1)

plot(sol, 0, 1, leg=false)
plot!(x, y, c=:red)
scatter!(x, y, c=:red, ms=4)
```

```julia
function euler_sistema(F, x0, W0, xf; n = 10)
    h = (xf - x0)/n
    x = linspace(x0, xf, n+1)
    m = length(W0)
    W = zeros(n+1, m)
    W[1, :] = W0
    for i = 1:n
        W[i+1,:] = W[i,:][:] + h * F(x[i], W[i,:])[:]
    end
    return x, W
end
```

```julia
g = 9.8
l = 4
F(x, Y) = [Y[2]; -(g/l) * sin(Y[1])]
W0 = [pi/4; 0.0]

t0 = 0
tf = 10;
x, W = euler_sistema(F, t0, W0, tf, n=1000);
```
