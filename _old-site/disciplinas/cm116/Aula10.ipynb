```julia
using Plots
```

```julia
# Retornando t e w para plotar
function euler_explicito(f, t0, y0, tf; n = 10)
    t = linspace(t0, tf, n+1) # ou t = 0.0:h:1.0
    if length(y0) == 1
        w = zeros(n+1) # Vetor
    else
        w = zeros(n+1,length(y0))
    end
    w[1,:] = y0
    h = (tf-t0)/n

    for i = 1:n
        w[i+1,:] = w[i,:][:] + h*f(t[i],w[i,:])
    end

    return t, w
end
```

```julia
# Sistema dinamico
# x' = -y
# y' = x
f(t,Y) =[-Y[2]; Y[1]]
Y0 = [1.0; 1.0]
t0 = 0
tf = 4π
N = 1000
t, W = euler_explicito(f, t0, Y0, tf; n = N);
plot(W[:,1], W[:,2], leg=false) #leg=false remove legenda
```

```julia
gr() # ou pyplot()
anim = Animation()
xmin = -2
xmax = 2
ymin = -2
ymax = 2
for j = 1:2:N
    scatter(W[j:j,1], W[j:j,2])
    plot!(W[max(1,j-10):j,1], W[max(1,j-10):j,2])
    xlims!(xmin, xmax)
    ylims!(ymin, ymax)
    frame(anim)
end
gif(anim, "tmp.gif", fps=20)
```
