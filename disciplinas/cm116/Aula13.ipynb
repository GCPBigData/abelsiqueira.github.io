```julia
function difusao_dissipacao_neumann(f, L, c, α, β; N = 10)
    h = L/(N-1)
    x = linspace(0, L, N)
    w = zeros(N)
    diagu = -[2.0; ones(N-2)]
    diagl = -[ones(N-2); 2.0]
    diag  = 2 + h^2*c(x)
    
    A = spdiagm( (diagl, diag, diagu), (-1,0,1))
    b = h^2*f(x)
    b[1] -= 2*h*α
    b[N] += 2*h*β
    w = A\b
    return x, w
end
```

```julia
using Plots
gr()
```

```julia
f(x) = 400*x.*(1-x)
c(x) = 1e1*ones(length(x))
α = 1
β = 0
L = 1
x, w = difusao_dissipacao_neumann(f, L, c, α, β, N = 100);
```

```julia
plot(x, w)
```

```julia
function adveccao_difusao_neumann(f, L, σ, c, α, β; N = 10)
    h = L/(N-1)
    x = linspace(0, L, N)
    w = zeros(N)
    diagu = -[2.0; 1 + h*σ(x[2:N-1])/2]
    diagl = -[1 - h*σ(x[2:N-1])/2; 2.0]
    diag  = 2 + h^2*c(x)
    
    A = spdiagm( (diagl, diag, diagu), (-1,0,1))
    b = h^2*f(x)
    b[1] -= (2 + h*σ(x[1])[1]) * h*α
    b[N] -= (-2 + h*σ(x[N])[1]) * h*β
    w = A\b
    return x, w
end
```

```julia
f(x) = 400*x.*(1-x)
c(x) = 1e1*ones(length(x))
σ(x) = -1e1*ones(length(x))
α = 1
β = 0
L = 1
x, w = adveccao_difusao_neumann(f, L, σ, c, α, β, N = 100);
```

```julia
plot(x, w)
```
