# Minimização em 1 variável

```julia
function bisseccao(f, a, b)
    tol = 1e-6
    x = (a+b)/2
    while abs(f(x)) > tol
        # Ou f(x) > 0 ou f(x) < 0
        if f(x) * f(a) < 0
            # O novo intervalo será [a,x]
            b = x
        else
            a = x
        end
        x = (a+b)/2
    end
    return x
end
```

```julia
f(x) = x^2 - 2
x = bisseccao(f, 1, 2)
println("x = $x, f(x) = $(f(x))")
```

```julia
function bisseccao(f, a, b; tol = 1e-6, maxiter = 1000)
    fa, fb = f(a), f(b)
    if abs(fa) <= tol
        b = a
    elseif abs(fb) <= tol
        a = b
    elseif fa * fb > 0
        error("Fi, não sei se tem raiz aí")
    end
    x = (a+b)/2
    fx = f(x)
    k = 0
    exitflag = 0 # 0 quer dizer ok
    while abs(fx) > tol || b-a > tol
        # Ou f(x) > 0 ou f(x) < 0
        if fx * fa < 0
            # O novo intervalo será [a,x]
            b, fb = x, fx
        else
            a, fa = x, fx
        end
        x = (a+b)/2
        fx = f(x)
        k = k + 1
        if k >= maxiter
            exitflag = 1
            break
        end
    end
    return x, fx, exitflag, k
end
```

```julia
f(x) = 1e-12*(x^2 - 2)
x, fx, ef, k = bisseccao(f, -1, 2, tol=1e-6)
println("x = $x, f(x) = $(fx)")
```

```julia
f(x) = x^2 + 1
x, fx, ef = bisseccao(f, 0, 1, tol=1e-6)
println("x = $x, f(x) = $(f(x)), ef = $ef")
```

```julia
f(x) = 1e24*(x^2 - 2)
x, fx, ef = bisseccao(f, -1, 2, tol=1e-2)
println("x = $x, f(x) = $(fx), ef = $ef")
```

```julia
f(x) = x^2 - 2
x, fx, ef, k = bisseccao(f, 0, 5, maxiter = 20)
println("x = $x, f(x) = $fx, ef = $ef, k = $k")
```

## Busca Áurea

```julia
function min1d(f, a, b; tol = 1e-6)
    phiinv = φ-1
    c = b - phiinv*(b-a)
    if f(c) > min(f(a),f(b))
        error("Deu ruim")
    end
    d = a + phiinv*(b-a)
    while b-a > tol
        if f(c) < f(d) # (a,c,d,b) <- (a,novo,c,d)
            d,b = c,d
            c = b - phiinv*(b-a)
        else # (a,c,d,b) <- (c,d,novo,b)
            a,c = c,d
            d = a + phiinv*(b-a)
        end
    end
    if f(c) < f(d)
        x, fx = c, f(c)
    else
        x, fx = d, f(d)
    end
    return x, fx
end
```

```julia
f(x) = (x - pi)^2 + e
x, fx = min1d(f, 2, 4)
println("x = $x, fx = $fx")
```
