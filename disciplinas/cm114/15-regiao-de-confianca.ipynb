```julia
B = [4.0 1.0; 1.0 2.0]
g = -B*[1.2; 0.8] # Roubei

m(d) = 0.5*dot(d, B*d) + dot(d, g)
```

```julia
contour(linspace(-0.2, 2, 100), linspace(-0.2, 2, 100),
(x,y)->m([x;y]))
```

```julia
d = -B\g
```

```julia
norm(d)
```

```julia
t = linspace(0, 2pi, 100)
@manipulate for Δ = linspace(0.1, 2, 100),
    λ = linspace(0, 10, 100)
    contour(linspace(-0.2, 2, 100), linspace(-0.2, 2, 100),
        (x,y)->m([x;y]))
    plot!(Δ*cos(t), Δ*sin(t), c=:blue, l=:dash)
    d = -(B + λ*eye(2))\g
    
    plot!([0, d[1]], [0, d[2]], c=:red)
    
    xlims!(-0.2, 2)
    ylims!(-0.2, 2)
end
```

```julia
function more_sorensen(B, g, Δ)
    n = length(g)
    d = -B\g
    if norm(d) < Δ
        return d
    end
    
    λlow, λ = 0.0, 1.0
    
    d = -(B + λ*eye(n))\g
    while norm(d) > Δ
        λlow = λ
        λ *= 10.0
        d = -(B + λ*eye(n))\g
    end
    
    λhigh = λ
    λ = (λlow + λhigh)/2
    d = -(B + λ*eye(n))\g
    while abs(norm(d) - Δ) > 1e-6
        if norm(d) > Δ
            λlow = λ
        else
            λhigh = λ
        end
        λ = (λlow + λhigh)/2
        d = -(B + λ*eye(n))\g
    end
    
    return d
end
```
