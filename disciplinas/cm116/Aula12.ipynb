```julia
f(x) = 400*x.*(1-x)
L = 1
h = 1/4
```

```julia
A = [-2  1  0; 1 -2  1; 0  1 -2]
b = -h^2*f([h; 2h; 3h]) - [20; 0; 20]
```

```julia
function eq_calor_contorno_fixo(f, L, T0, TL; N = 10)
    h = L/(N-1)
    x = linspace(0, L, N)
    w = zeros(N)
    w[1] = T0
    w[N] = TL
    A = spdiagm((ones(N-3),-2*ones(N-2),ones(N-3)), (-1,0,1))
    b = -h^2*f(x[2:N-1])
    b[1] -= T0
    b[N-2] -= TL
    w[2:N-1] = A\b
    return x, w
end
```

```julia
T0 = 20
TL = 20
x, w = eq_calor_contorno_fixo(f, L, T0, TL)
```

```julia
plot(x, w)
plot!(sol, 0, L)
```

```julia
L = 1
f(x) = 1000*exp( 5*(x-L/2).^2 )
T0 = 0
TF = 100
x, w = eq_calor_contorno_fixo(f, L, T0, TF, N = 100)
plot(x,w)
```
