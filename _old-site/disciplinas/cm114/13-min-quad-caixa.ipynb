$$ x_i - u_i = tg_i $$

$$ t = \frac{x_i - u_i}{g_i} $$

```julia
g = ∇f(x0)
steps = Float64[]
for i = 1:2
    if g[i] < 0 && u[i] < Inf
        push!(steps, (x0[i] - u[i])/g[i])
    elseif g[i] > 0 && l[i] > -Inf
        push!(steps, (x0[i] - l[i])/g[i])
    end
end
steps = sort(steps)
println("steps = $steps")
```

```julia
using ForwardDiff

function gradient_method(f, x, l, u; tol = 1e-5, max_iter = 1000, max_time = 60)
    exit_flag = 0
    ∇f(x) = ForwardDiff.gradient(f, x)
    
    iter = 0
    start_time = time()
    elapsed_time = 0.0
    
    P(x, l, u) = min(u, max(l, x))
    x = P(x, l, u)
    fx = f(x)
    ∇fx = ∇f(x)
    while norm(P(x-∇fx,l,u)-x) > tol
        t = 1.0
        xp = P(x - t*∇fx, l, u)
        while f(xp) > fx + 0.5*t*dot(∇fx, xp - x)
            t = t*0.9
            xp = P(x - t*∇fx, l, u)
            if t < 1e-14
                exit_flag = -1
            end
        end
        if exit_flag == -1
            break
        end
        x = xp
        fx = f(x)
        ∇fx = ∇f(x)
        iter = iter + 1
        if iter >= max_iter
            exit_flag = 1
            break
        end
        elapsed_time = time() - start_time
        if elapsed_time >= max_time
            exit_flag = 2
            break
        end
    end
    return x, fx, ∇fx, exit_flag, iter, elapsed_time # Precisamos retornar o ponto encontrado
end
```

```julia
f(x) = (x[1] - 1)^2 + 4*(x[2] - 1)^2
x0 = zeros(2)
l = [1.1; 0.9]
u = [1.9; 1.1]
gradient_method(f, x0, l, u)
```

```julia
f(x) = (x[1] - 1)^2 + 100*(x[2] - x[1]^2)^2
x0 = zeros(2)

u = [1.3; 1.9]
gradient_method(f, x0, l, u, max_iter=100000)
```
