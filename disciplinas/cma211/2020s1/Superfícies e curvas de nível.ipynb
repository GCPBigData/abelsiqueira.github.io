# Superfícies e curvas de nível

```julia
lay = @layout [a b]

f(x,y) = x^2 + y^2
xg = range(-2, 2, length=100)
yg = range(-2, 2, length=100)
p = plot(layout=lay, size=(800,300))
surface!(p[1], xg, yg, f)
contour!(p[2], xg, yg, f)
```

```julia
lay = @layout [a b]

f(x,y) = -2x^2 - y^2
xg = range(-2, 2, length=100)
yg = range(-2, 2, length=100)
p = plot(layout=lay, size=(800,300))
surface!(p[1], xg, yg, f)
contour!(p[2], xg, yg, f)
```

```julia
lay = @layout [a b]

f(x,y) = x^2 - y^2
xg = range(-2, 2, length=100)
yg = range(-2, 2, length=100)
p = plot(layout=lay, size=(800,300))
surface!(p[1], xg, yg, f)
contour!(p[2], xg, yg, f)
```

```julia
lay = @layout [a b]

f(x,y) = x^3 - 3x - y^2
xg = range(-2, 2, length=100)
yg = range(-2, 2, length=100)
p = plot(layout=lay, size=(800,300))
surface!(p[1], xg, yg, f, camera=(70,40))
contour!(p[2], xg, yg, f)
```

```julia
lay = @layout [a b]

f(x,y) = (x^2 - y^2) / (x^2 + y^2)
xg = range(-2, 2, length=100)
yg = range(-2, 2, length=100)
p = plot(layout=lay, size=(800,300))
surface!(p[1], xg, yg, f, camera=(30,70))
contour!(p[2], xg, yg, f)
```

```julia
lay = @layout [a b]

f(x,y) = x^2 * y / (x^4 + y^2)
xg = range(-2, 2, length=100)
yg = range(-2, 2, length=100)
p = plot(layout=lay, size=(800,300))
surface!(p[1], xg, yg, f)
contour!(p[2], xg, yg, f)
```

```julia
lay = @layout [a b]

f(x,y) = sqrt((x - 1)^2 + 4 * (y - x^2)^2)
xg = range(-2, 2, length=100)
yg = range(-2, 2, length=100)
p = plot(layout=lay, size=(800,300))
surface!(p[1], xg, yg, f)
contour!(p[2], xg, yg, f)
```

```julia
lay = @layout [a b]

f(x,y) = y * exp(-x)
xg = range(-2, 2, length=100)
yg = range(-2, 2, length=100)
p = plot(layout=lay, size=(800,300))
surface!(p[1], xg, yg, f)
contour!(p[2], xg, yg, f, levels=50)
```

```julia
lay = @layout [a b]

f(x,y) = sin(4x) * cos(4y) + 0.4*(x^2 + y^2)
xg = range(-2, 2, length=100)
yg = range(-2, 2, length=100)
p = plot(layout=lay, size=(800,300))
surface!(p[1], xg, yg, f)
contour!(p[2], xg, yg, f, levels=50)
```
