```julia
function difdiv(x, y)
    m = length(x)
    D = zeros(m, m)
    D[:,1] = y
    for j = 2:m
        for i = 1:m-j+1
            D[i,j] = (D[i+1,j-1] - D[i,j-1])/(x[i+j-1]-x[i])
        end
    end
    return D[1,:]
end
```

```julia
function difdiv_eval(t, a, x)
    m = length(x)
    p = a[m]
    for i = m-1:-1:1
        p = a[i] + p*(t - x[i])
    end
    return p
end
```

```julia
using Plots
gr()

x = collect(0:10)
y = sin(x) + 2*exp(-x/3) + sqrt(x)
scatter(x, y)
a = difdiv(x, y)
plot!(t->difdiv_eval(t, a, x), x[1], x[end])
```

## Aproximação de funções por polinômios

Dada $f:[a,b] \to \mathbb{R}$, queremos $p$, um polinômio, que aproxime $f$.

Uma alternativa é fazer **Quadrados Mínimos Contínuo**, opção bastante interessante, que não vamos ver aqui.

Outra é escolher um conjunto $m$ de pontos em $[a,b]$ e retornar a interpolação de $(x_i, f(x_i))$ para esses pontos.

```julia
function interpola(x, f::Function)
    y = map(f, x)
    a = difdiv(x, y)
    return t -> difdiv_eval(t, a, x)
end
```

```julia
x = linspace(-1, 1, 50)
f(x) = exp(x)
p = interpola(x, f)
plot(f, x[1], x[end], leg=false)
scatter!(x, map(f, x), ms=4, c=:blue)
plot!(p, x[1], x[end], c=:red)
```

**Pergunta:** Se $p_m$ é o polinômio que interpola $f$ em $[a,b]$ com $m$ pontos, então
$p_m \to f$ quando $m \to \infty$?

```julia
using Interact

f(x) = sin(x) + cos(2x)
a, b = -5, 5
@manipulate for m = slider(3:50, value=3)
    x = linspace(a, b, m)
    p = interpola(x, f)
    plot(f, x[1], x[end], leg=false)
    scatter!(x, map(f, x), ms=4, c=:blue)
    plot!(p, x[1], x[end], c=:red)
end
```

```julia
f(x) = 1/(1 + x^2)
a, b = -5, 5
@manipulate for m = slider(3:50, value=3)
    x = linspace(a, b, m)
    p = interpola(x, f)
    plot(f, x[1], x[end], leg=false)
    scatter!(x, map(f, x), ms=4, c=:blue)
    plot!(p, x[1], x[end], c=:red)
end
```

Esse efeito é conhecido como **Fenômeno de Runge**. Uma alternativa para resolvê-lo é
utilizar as **nós de Chebyshev**:
$$ t_i = \cos\bigg(\frac{(2i-1)\pi}{2m}\bigg). $$

Note que $t_i \in (-1,1)$. Para transladá-los para $[a,b]$, devemos fazer
$$ x_i = \frac{a+b}{2} + \frac{b-a}{2}t_i. $$

```julia
f(x) = 1/(1 + x^2)
a, b = -5, 5

@manipulate for m = slider(3:50, value=3)
    x = (a+b)/2 + (b-a)*cos( (2*(1:m) - 1)*π/2m )/2
    p = interpola(x, f)
    plot(f, a, b, leg=false)
    scatter!(x, map(f, x), ms=4, c=:blue)
    plot!(p, a, b, c=:red)
end
```

```julia
f(x) = sin(3x) + cos(5x)
a, b = -5, 5

@manipulate for m = slider(3:50, value=3)
    x = (a+b)/2 + (b-a)*cos( (2*(1:m) - 1)*π/2m )/2
    p = interpola(x, f)
    plot(f, a, b, leg=false)
    scatter!(x, map(f, x), ms=4, c=:blue)
    plot!(p, a, b, c=:red)
end
```

Os nós de Chebyshev são provenientes do polinômio de Chebyshev, e com eles podemos sim afirmar que $p_m \to f$.

# Splines

Outra maneira de encontrar uma interpolação para $(x_i,y_i), i = 1,\dots,m$, que não
envolve um modelo $f(x) = \alpha_1 \phi_1 + \dots + \alpha_n \phi_n$ são as Splines.

Splines são funções por parte, com o objetivo de ser fácil de calcular e guardar, principalmente quando $m$ é grande.

Por exemplo, o método das diferenças divididas precisa de uma quantidade de operações proporcionais a $m$ para calcular $p(x)$. Quando $m$ é muito grande, isso é muito caro.
As splines terão um custo constante no cálculo de $p(x)$.

Hipótese: $x_1 < x_2 < \dots < x_m$.

De uma maneira geral, uma spline que interpola $(x_i, y_i)$, $i = 1,\dots,m$ é dada por
$$ f(x) = \left\{\begin{array}{ll}
s_1(x), & x_1 \leq x \leq x_2, \\
s_2(x), & x_2 < x \leq x_3, \\
s_3(x), & x_3 < x \leq x_4, \\
\vdots & \vdots \\
s_{m-1}(x), & x_{m-1} < x \leq x_m, \\
\end{array}\right.$$
onde
1. $s_i(x_i) = y_i$, $i = 1,\dots,m-1$,
2. $s_i(x)$, $i=1,\dots,m$ são funções polinomiais de mesmo grau,
3. $s_i(x_{i+1}) = s_{i+1}(x_{i+1})$, $i = 1,\dots,m-1$,
4. Se o grau de $s_i$ é maior que 1, então,
$s_i^{(k)}(x_{i+1}) = s_{i+1}^{(k)}(x_{i+1})$, para todo $k = 1,\dots,gr(s_i)-1$,
$i = 2,\dots,m-1$.

A ordem de uma spline é o grau de $s_i$.

A spline de ordem 1 (linear) é simplesmente uma função linear por partes, "costurando" nos pontos $(x_i,y_i)$.

A spline de ordem 2 (quadrática) é uma função quadrática por partes diferenciável em $x_i$, $i = 2,\dots,m-1$.

A spline de ordem 3 (cúbica) é uma função cúbica por partes diferenciável até segunda ordem em $x_i$, $i = 2,\dots,m-1$.

E assim por diante.

De maneira geral, podemos escrever

$$ s_i(x) = \alpha_{0,i} + \alpha_{1,i}(x - x_i) + \dots + \alpha_{n,i}(x-x_i)^n, $$
onde $n$ é o grau da spline. Note que uma spline tem $m-1$ termos e $n+1$ incógnitas por termo, isto é, $(m-1)(n+1)$ incógnitas.
A condição 1 nos dá $m-1$ restrições, e a condição 3 nos dá outras $m-1$ restrições.
Para uma spline linear, as incógnitas são 2(m-1), então ela é unicamente determinada.

Para splines de ordem maior, a condição 4 dá $m-2$ restrições por grau acima do primeiro, enquanto que o número de incógnitas cresce em $m-2$ por grau.
Isso quer dizer que a spline quadrática precisa de mais uma restrição, que a cúbica precisa de mais duas, e assim por diante. Existem mais de uma maneira de lidar com isso.

Obs.: Se tratando de expandir o intervalo de interesse, podemos extender uma spline fora
do intervalo $[x_1,x_m]$ usando $s(x) = s_1(x)$ se $x < x_1$ e $s(x) = s_{m-1}(x)$ se $x > x_m$.

## Spline Linear

Na spline linear, cada $s_i(x)$ pode ser escrito como
$$ s_i(x) = a_i + b_i(x - x_i). $$

Agora, a equação 1
$$ s_i(x_i) = y_i \qquad \Rightarrow \qquad
a_i = y_i. $$
E a equação 3 nos dá
$$ s_i(x_{i+1}) = y_{i+1} \qquad \Rightarrow \qquad
b_i = \frac{y_{i+1}-y_i}{x_{i+1}-x_i}. $$

```julia
function spline_lin(x, y)
    m = length(x)
    P = zeros(m-1, 2) # P = [a1 b1; a2 b2; ...]
    P[:,1] = y[1:m-1]
    P[:,2] = (y[2:m] - y[1:m-1])./(x[2:m] - x[1:m-1])
    return P
end
```

```julia
function spline_lin_eval(t, P, x)
    # Busca do intervalo
    m = length(x)
    j = 1
    if t > x[m]
        j = m-1
    else
        while t > x[j+1]
            j += 1
        end
    end
    return P[j,1] + P[j,2] * (t - x[j])
end
```

```julia
a, b = -5, 5
@manipulate for m = 3:50
    x = linspace(a, b, m)
    f(x) = 1/(1 + x^2)
    y = map(f, x)
    P = spline_lin(x, y)

    scatter(x, y, ms=4, leg=false)
    plot!(f, a, b)
    plot!(t->spline_lin_eval(t, P, x), a, b)
end
```
