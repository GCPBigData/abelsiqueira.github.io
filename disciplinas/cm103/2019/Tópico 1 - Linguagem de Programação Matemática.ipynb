**Bháskara:**
$ax^2 + bx + c = 0$ e $b^2 - 4ac \geq 0$ implicam em
$x = \dfrac{-b \pm \sqrt{b^2 - 4ac}}{2a}. $

`````julia
function bhaskara(a, b, c)
    Δ = b^2 - 4 * a * c
    if Δ < 0
        error("Nos reais não tem solução")
    end
    return (-b + sqrt(Δ)) / 2a, (-b - sqrt(Δ)) / 2a
end
`````

`````julia
bhaskara(1, 0, -1)
`````

`````julia
bhaskara(1, 0, 1)
`````

`````julia
bhaskara(1.0, 5.0, 6.0)
`````

`````julia
bhaskara(0, 1, 1)
`````

**Maior potência de 2 menor ou igual que um número**

`````julia
function maior_pot(x)
    p = 1
    while p <= x
        p = 2p
    end
    p = div(p, 2)
    return p
end
`````

`````julia
maior_pot(2)
`````

`````julia
maior_pot(10)
`````

`````julia
maior_pot(100)
`````

`````julia
for n = [1, 10, 100, 1000, 10000, 100000, 1000000]
    p = maior_pot(n)
    println("2ᵏ = p ≤ n, n = $n, p = $p")
end
`````

**MMC:**
$\mbox{mmc}(a, b)$ é o menor número natural que é múltiplo de $a$ e de $b$.

    or -> ||
    and -> &&

`````julia
function mmc(a::Int, b::Int)
    if a < 1 || b < 1
        error("Entrada deve ser de dois inteiros positivos")
    end
    xa, xb = a, b # Múltiplos de a e b
    while (xa != xb) # Não preciso dos parenteses
        if xa < xb
            xa = xa + a
        else
            xb = xb + b
        end
    end
    return xa
end
`````

`````julia
mmc(2,3)
`````

`````julia
mmc(5, 7)
`````

`````julia
mmc(6, 8)
`````

`````julia
mmc(12, 14)
`````

`````julia
mmc(-1, 0)
`````

`````julia
mmc(2.0, 3.0)
`````

`````julia
function eh_primo(n :: Int)
    if n == 1
        return false
    elseif n == 2
        return true
    elseif n % 2 == 0
        return false
    end
    for k = 3:2:ceil(Int, sqrt(n))
        if n % k == 0
            return false
        end
    end
    return true
end
`````

`````julia
u0(x, y) = exp(-((x - y^2)^2 + x^2))
X = -1.5:0.1:2.5
Y = -2:0.1:2

S = sum(u0.(X, Y)) * (X[2] - X[1]) * (Y[2] - Y[1])
u(t,x,y) = exp(-t) * (t * S + u0(x,y) * (1 - t))

t = 0:0.01:1
anim = Animation()
for ti = t
    wireframe(X, Y, (x,y) -> u(ti,x,y), leg=false)
    zlims!(0, 1)
    frame(anim)
end
gif(anim, "exemplo5.gif", fps=12)
`````

`````julia
gr()

x = range(0, stop=2*pi, length=100)
@manipulate for a = x
    plot(x, sin.(x), leg=false)
    scatter!([a], [sin(a)], c=:red)
    plot!(x, sin(a) .+ cos(a) * (x .- a) .- sin(a) * (x .- a).^2 / 2, c=:red, l=:dash)
    xlims!(x[1], x[end])
    ylims!(-2, 2)
end
`````

`````julia
gr(size=(400,400))
@manipulate for N = slider(0:5, value=0)
    θ = [k * 2π / 3 for k = 0:3]
    P = [cos.(θ) sin.(θ)]' # 2 × 4
    
    for t = 1:N
        P_new = zeros(2, 0)
        n = size(P, 2)
        for i = 1:n-1
            x, y = P[:,i], P[:,i+1]
            v = (y - x) / 3
            vn = sqrt(3) / 2 * [v[2]; -v[1]] # 90ᵒ ↶
            P_new = [P_new  x  x + v  x + 1.5v + vn  x + 2v] # Ultimo ponto não entra
        end
        P_new = [P_new P[:,n]]
        P = P_new
    end
    
    plot(P[1,:], P[2,:], c=:blue, lw=2, leg=false, axis=false, grid=false)
    xlims!(-1, 1)
    ylims!(-1, 1)
end
`````
