## Integração Numérica

Em cálculo I aprendemos que
$$ \int_a^b f(x)\mbox{d}x = F(b) - F(a), $$
se $f$ for uma função integrável, e onde $F$ é uma primitiva de $f$.

No entanto, nem sempre é fácil encontrar uma primitiva para $F$.
Às vezes, pode ser impossível, como em
$$ \int_a^b e^{-x^2}\mbox{d}x. $$

Por sorte, muitas vezes uma aproximação numérica é suficiente para nosso objetivo.

Vamos estudar alguns métodos de aproximação numérica, a maior parte deles usando valores da
função $f$ em pontos selecionados, isto é
$$ \int_a^bf(x)\mbox{d}x = \sum_{i=1}^m A_if(x_i), $$
para uma quantidade de pontos $m$, pontos $x_i$ e valores $A_i$.

Salvo indicação contrária, iremos supor que $x_1 < x_2 < \dots < x_m$.

## Fórmulas de Newton-Cotes

As fórmulas de Newton-Cotes provém da interpolação da função $f$ em pontos igualmente espaços do
intervalo $[a,b]$. Se os intervalos estiverem entre os pontos escolhidos, isto é, se $x_1 = a$ e
$x_m = b$, então a fórmula é dita **fechada**. Caso contrário, aberta.

Vamos começar considerando uma simples aproximação para $f$: uma interpolação de um ponto.
$$ p(x) = f\bigg(\frac{a+b}{2}\bigg), $$
isto é, $p$ é uma função constante. Essa aproximação resulta em
$$ \int_a^b f(x)\mbox{d}x \approx \int_a^bf((a+b)/2)\mbox{d}x = (b-a)f\bigg(\frac{a+b}{2}\bigg). $$

Essa aproximação é chamada de regra do **ponto-médio**, e é uma fórmula aberta.

```julia
using Plots
gr()
```

```julia
f(x) = exp(x)
F(x) = exp(x) # Primitiva para fins de comparação
a, b = -1, 1
plot(f, a, b, c=:red, fill=(0,:red,0.5), leg=false)
plot!(x->f((a+b)/2), a, b, c=:blue, fill=(0,:blue,0.5))
```

```julia
function ponto_medio(f, a, b)
    return (b-a)*f((a+b)/2)
end
```

```julia
ponto_medio(f, a, b) - (F(b)-F(a))
```

Agora uma aproximação com dois pontos.
$$ p(x) = f(x_1) + (x-x_1)\frac{f(x_2)-f(x_1)}{x_2-x_1}. $$

Podemos escolher os pontos de duas maneiras: $x_1 = a$ e $x_2 = b$, ou de modo que $a, x_1, x_2, b$ sejam igualmente espaçados. Curiosamente, as duas maneiras nos dão
$$ \int_a^bf(x)\mbox{d}x \approx \frac{b-a}{2}[f(x_1) + f(x_2)], $$
porém com $x_1$ e $x_2$ diferentes. No caso aberto, temos $x_1 = a + (b-a)/3 = (2a+b)/3$ e $x_2 = a + 2(b-a)/3 = (a+2b)/3$.

O caso fechado é chamado tradicionalmente de regra do trapézio, mas os dois formas trapézios, então vamos diferenciá-los por regra do trapézio fechada e regra do trapézio aberta.

```julia
plot(f, a, b, c=:red, fill=(0,:red,0.5), leg=false)
plot!(x->f(a) + (x-a)*(f(b)-f(a))/(b-a), a, b, c=:blue, fill=(0,:blue,0.5))
```

```julia
plot(f, a, b, c=:red, fill=(0,:red,0.5), leg=false)
x1, x2 = (2a+b)/3, (a+2b)/3
plot!(x->f(x1) + (x-x1)*(f(x2)-f(x1))/(x2-x1), a, b, c=:blue, fill=(0,:blue,0.5))
```

```julia
function trapezio_fechado(f, a, b)
    return (b-a)*(f(a)+f(b))/2
end

function trapezio_aberto(f, a, b)
    h = (b-a)/3
    return (b-a)*(f(a+h)+f(b-h))/2
end
```

```julia
trapezio_fechado(f, a, b) - (F(b)-F(a))
```

```julia
trapezio_aberto(f, a, b) - (F(b)-F(a))
```

Um aspecto não explorado em interpolação foi o erro. Aqui em integração, também temos um erro envolvido. Como não vimos antes, agora iremos direto ao resultado.

As três fórmulas apresentadas têm o mesma ordem de erro, que é proporcional a $(b-a)^3f''(\xi)$ para algum ponto $\xi\in[a,b]$. Isso quer dizer que as três são proporcionais em termo de convergência. Isso quer dizer também que entre as duas abertas, a do ponto-médio é melhor, porque tem uma avaliação a menos de função.

A ideia agora seria aumentar o grau da interpolação. Porém, antes de fazer isso, vamos lembrar que se o método é baseado em interpolação usando intervalos igualmente espaçados, então o fenômeno de Runge é uma possibilidade.
Dessa maneira, vamos fazer uma pequena alteração, e quebrar o intervalo $[a,b]$ em vários intervalos menores.
Em cada um desses intervalos vamos fazer a aplicação do método, obtendo uma **regra composta**.

```julia
using Interact
```

```julia
f(x) = sin(4x) + cos(5x)
F(x) = -cos(4x)/4 + sin(5x)/5
@manipulate for m = slider(1:100, value=1)
    h = (b-a)/m
    plot(f, a, b, c=:red, fill=(0,:red,0.5), leg=false)    
    for i = 1:m
        x1 = a + i*h - h/2
        plot!(x->f(x1), x1-h/2, x1+h/2, c=:blue, fill=(0,:blue,0.5))
    end
    plot!()
end
```

```julia
function ponto_medio_composto(f, a, b; m = 10)
    I = 0.0
    h = (b-a)/m
    for i = 1:m
        I += f(a + i*h - h/2)
    end
    return I * h
end
```

```julia
ponto_medio_composto(f, a, b, m=100) - (F(b)-F(a))
```

```julia
Erro = map(m->ponto_medio_composto(f, a, b, m=m)-(F(b)-F(a)), 1:100)
plot(1:100, abs(Erro), yaxis=:log)
```

```julia
@manipulate for m = slider(2:100, value=2)
    h = (b-a)/(m-1)
    plot(f, a, b, c=:red, fill=(0,:red,0.5), leg=false)    
    for i = 1:m-1
        x1, x2 = a + (i-1)*h, a + i*h
        plot!(x->f(x1) + (x-x1)*(f(x2)-f(x1))/(x2-x1), x1, x2, c=:blue, fill=(0,:blue,0.5))
    end
    plot!()
end
```

No caso do método de trapézio, temos
$$ \int_a^bf(x)\mbox{d}x =
\sum_{i=1}^{m-1}\int_{x_i}^{x_{i+1}}f(x)\mbox{d}x
\approx \sum_{i=1}^{m-1}\frac{h[f(x_i)+f(x_{i+1})]}{2}
= \frac{h}{2}\bigg[f(a)+f(b) + 2\sum_{i=2}^{m-1} f(x_i)\bigg].
$$

```julia
function trapezio_composto(f, a, b; m = 10)
    if m < 2
        error("m deve ser maior ou igual a 2")
    end
    I = f(a) + f(b)
    h = (b-a)/(m-1)
    for i = 1:m-2
        I += 2*f(a + i*h)
    end
    return I * h / 2
end
```

```julia
trapezio_composto(f, a, b, m=100) - (F(b)-F(a))
```

```julia
Erro = map(m->trapezio_composto(f, a, b, m=m)-(F(b)-F(a)), 2:100)
plot(2:100, abs(Erro), yaxis=:log)
```

```julia
Erro = map(m->ponto_medio_composto(f, a, b, m=m)-(F(b)-F(a)), 2:100)
plot!(2:100, abs(Erro), yaxis=:log, c=:red)
```

A próxima fórmula, provavelmente a mais comum, é obtida interpolando a função $f$ em três pontos: $a$, $(a+b)/2$ e $b$. Escrevendo $h = (b-a)/2$, e podemos fazer a interpolação de Lagrange em $a$, $a+h$ e $a+2h$.
