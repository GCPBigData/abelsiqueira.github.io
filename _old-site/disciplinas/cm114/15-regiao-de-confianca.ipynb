```julia
B = [4.0 1.0; 1.0 2.0]
g = -B*[1.2; 0.8] # Roubei

m(d) = 0.5*dot(d, B*d) + dot(d, g)
```

```julia
contour(linspace(-0.2, 2, 100), linspace(-0.2, 2, 100),
(x,y)->m([x;y]))
```

```julia
d = -B\g
```

```julia
norm(d)
```

```julia
t = linspace(0, 2pi, 100)
@manipulate for Δ = linspace(0.1, 2, 100),
    λ = linspace(0, 10, 100)
    contour(linspace(-0.2, 2, 100), linspace(-0.2, 2, 100),
        (x,y)->m([x;y]))
    plot!(Δ*cos(t), Δ*sin(t), c=:blue, l=:dash)
    d = -(B + λ*eye(2))\g
    
    plot!([0, d[1]], [0, d[2]], c=:red)
    
    xlims!(-0.2, 2)
    ylims!(-0.2, 2)
end
```

```julia
function more_sorensen(B, g, Δ)
    n = length(g)
    d = -B\g
    if norm(d) < Δ
        return d
    end
    
    λlow, λ = 0.0, 1.0
    
    d = -(B + λ*eye(n))\g
    while norm(d) > Δ
        λlow = λ
        λ *= 10.0
        d = -(B + λ*eye(n))\g
    end
    
    λhigh = λ
    λ = (λlow + λhigh)/2
    d = -(B + λ*eye(n))\g
    while abs(norm(d) - Δ) > 1e-6
        if norm(d) > Δ
            λlow = λ
        else
            λhigh = λ
        end
        λ = (λlow + λhigh)/2
        d = -(B + λ*eye(n))\g
    end
    
    return d
end
```

```julia
more_sorensen(B, g, 1.0)
```

```julia
using Krylov

t = linspace(0, 2pi, 100)
@manipulate for Δ = linspace(0.1, 2, 1000)
    contour(linspace(-0.2, 2, 100), linspace(-0.2, 2, 100),
        (x,y)->m([x;y]), leg=false)
    plot!(Δ*cos(t), Δ*sin(t), c=:blue, l=:dash)
    
    # Gradiente
    plot!([0, -g[1]], [0, -g[2]], c=:blue)
    λ = dot(g,g)/dot(g,B*g)
    dC = -λ*g
    dN = -B\g
    scatter!([dC[1], dN[1]], [dC[2], dN[2]], c=:magenta, ms=4)
    
    d = more_sorensen(B, g, Δ)
    
    plot!([0, d[1]], [0, d[2]], c=:red)
    
    scg, _ = Krylov.cg(B, -g, radius=Δ)
    plot!([0, scg[1]], [0, scg[2]], c=:green, lw=2, l=:dash)
    
    xlims!(-0.2, 2)
    ylims!(-0.2, 2)
end
```
