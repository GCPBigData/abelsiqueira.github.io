# Condição de Wolfe

Como vimos, dada uma direção $d^k$ a partir de $x^k$, a busca linear consiste de encontrar um tamanho de passo $t_k$ tal que $x^{k+1} = x^k + t_kd^k$ seja bom.
Já vimos a busca exata e a condição de Armijo. Agora vamos ver a condição de Wolfe.

A condição de Wolfe consiste em escolher um tamanho $t_k$ tal que
$$f(x^k + td^k)  \leq f(x^k) + c_1t_k\nabla f(x^k)^Td^k
\qquad \mbox{e} \qquad
\nabla f(x^k + td^k)^Td^k \geq c_2 \nabla f(x^k)^Td^k $$
Se definimos $\phi(t) = f(x^k + td^k)$, essas condições podem ser escritas como
$$ \phi(t) \leq \phi(0) + c_1t_k\phi'(0)
\qquad \mbox{e} \qquad
\phi'(t) \geq c_2\phi'(0). $$
A primeira condição é a mesma que Armijo, e a segunda dá uma condição sobre a ângulo da direção.

A primeira condição pode ser satisfeita como fizemos com Armijo. Se fazemos um decréscimo suficiente, então encontraremos um passo.
A segunda condição, no entanto, não é tão simples.

Na figura abaixo mostramos um exemplo.

```julia
using Winston
x = 1.0
f(x) = x.^8/8; ∇f(x) = x.^7
d = -0.005 # O passo é muito pequeno
ϕ(t) = f(x + t*d)
ϕd(t) = d*∇f(x+t*d)
t = linspace(0,1,100);
Winston.plot(t, ϕ(t), t, ϕ(0) + 1e-4*t*ϕd(0), "r") # O decréscimo acontece pra qualquer tamanho de passo 0 ≦ t ≦ 1.
```

```julia
Winston.plot(t, ϕd(t), t, 0.9*ϕd(0)*ones(t),"r") # Mas a condição da derivada não acontece para nenhum dele
```

```julia
t = linspace(0,10,100);
Winston.plot(t, ϕd(t), t, 0.9*ϕd(0)*ones(t),"r")
```
